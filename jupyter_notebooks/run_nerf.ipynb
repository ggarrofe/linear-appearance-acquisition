# Residual Reflectance

```python
import torch
values = torch.tensor([[1.2, 2.1, 3.2], [1.4, 2.2, 3.1], [1.5, 2.2, 3.1]])
samples = torch.tensor([[2.1, 0.2, 1.3], [2.9,3.2,5.3], [.12, 12.2, 2.4]])
samplesA = samples.unsqueeze(1)
samplesB = samples.unsqueeze(0)
dif = samplesA - samplesB
print(dif)
print(dif.shape)
dist = torch.norm(dif, dim=-1)
print(dist)
print(dist > 6)
mask = torch.triu(dist > 6, diagonal=1)
indices = torch.argwhere(mask)
print(indices)
print(indices[:,0])
print(values[indices[:,0]])
print(values[indices[:,1]])
loss = values[indices[:,0]] - values[indices[:,1]]
print(loss)
loss_fn = torch.nn.MSELoss(reduction='none')
print(loss_fn(values[indices[:,0]], values[indices[:,1]]))
loss_fn = torch.nn.MSELoss(reduction='sum')
print(loss_fn(values[indices[:,0]], values[indices[:,1]]))
loss_fn = torch.nn.MSELoss(reduction='mean')
print(loss_fn(values[indices[:,0]], values[indices[:,1]]))
```
